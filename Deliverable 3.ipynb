# Fitting the Linear Regression Model

```python
#importing the linear model package
from sklearn.linear_model import LinearRegression
```

```python
#Seprating the dependent and independent variables
Y=data[['predrt']]
X=data[['perc_non_white']]
```

```python
#X.shape
```

```python
#X.head(1)
```

```python
#Y.shape
```

```python
#Y.head(1)
```

```python
Y.isnull().sum()
```

```python
M1=LinearRegression()
```

```python
#Fitting the Linear Regression Model on the Data
M1.fit(X,Y)
```

```python
 print('Coefficients: \n', M1.coef_)
```

```python
 print('Coefficients: \n', M1.intercept_)
```

```python
#Predicting the values of Y
M1_pred=M1.predict(X)
```

```python
#importing the package to get r2
from sklearn.metrics import r2_score
```

```python
R2=r2_score(Y, M1_pred)
```

```python
#The R2 value for the model was 14%
R2*100
```

```python
#Importing the stats model package
import statsmodels.api as sm
```

```python
#printing the summary of the regression model
X1= sm.add_constant(X)
est = sm.OLS(Y, X1)
est2 = est.fit()
print(est2.summary())
```
